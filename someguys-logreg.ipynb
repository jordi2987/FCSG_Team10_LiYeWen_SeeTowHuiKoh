```python
df = pd.read_csv("data.csv")
df.head()
```

```python
#TF-IDF: convert the text data into a format suitable for ML models.
from sklearn.feature_extraction.text import TfidfVectorizer

#initialise TF-IDF Vectorizer, considering the top 1000 words and convert to uni-grams and bigrams, and remove stopwords
vectorizer = TfidfVectorizer(max_features=1000, ngram_range = (1,2), stop_words='english')
```

```python
#apply TF-TDF to combined text data, Term-Frequency emphasizes frequent words, Inverse-Document-Frequency reduces weight of common words
tfidf_matrix = vectorizer.fit_transform(df['combined'])

#check the shape of the matrix
tfidf_matrix.shape
```

```python
#creating a dataframe for the TF-IDF values
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns = vectorizer.get_feature_names_out())

#show the dataframe
tfidf_df.head()
```

3a. Sentiment analysis

```python
#pip install textblob
from textblob import TextBlob

def get_sentiment(text):
  sentiment = TextBlob(text).sentiment.polarity
  return sentiment

#
df['sentiment_score'] = df['combined'].apply(get_sentiment)
df['sentiment_score'].head()
```

3b. Historical Price Features

```python
df_prices = pd.read_csv("upload_DJIA_table.csv")

df_prices['Date'] = pd.to_datetime(df_prices['Date'])
df.sort_values('Date', ascending=True, inplace=True)
df['7_day_avg'] = df_prices['Close'].rolling(window=7).mean()
df['30_day_avg'] = df_prices['Close'].rolling(window=30).mean()
df['daily_return'] = df_prices['Close'].pct_change(1)
df['volatility_30'] = df['daily_return'].rolling(window=30).std()
```

Linear Regression: this is where the funs starts

```python
# For time-series data, we have to consider the time-sensitive context of the data. 
# Therefore, we use a train-test split of
df.dropna(inplace=True)

split_point = int(len(df) * 0.8)

train_df = df.iloc[:split_point]
test_df = df.iloc[split_point:]
#test_df.head()
```

```python
X_train = train_df[['sentiment_score', '7_day_avg', '30_day_avg', 'daily_return', 'volatility_30']]
X_train.info()
y_train = train_df['Label']
X_test = test_df[['sentiment_score', '7_day_avg', '30_day_avg', 'daily_return', 'volatility_30']]
y_test = test_df['Label']
```

Plot graphs of sentiment socre, 7 day avg, 30 day avg, daily return, volatility 30

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score


# Extract features and target variable for both training and test sets

X_train = train_df[['sentiment_score', '7_day_avg', '30_day_avg', 'daily_return', 'volatility_30']]
y_train = train_df['Label']
X_test = test_df[['sentiment_score', '7_day_avg', '30_day_avg', 'daily_return', 'volatility_30']]
y_test = test_df['Label']

# Initialize and train the Logistic Regression model
model = LogisticRegression()
model.fit(X_train, y_train)

# Predictions
predictions = model.predict(X_test)

# Evaluate the model
print("Accuracy:", accuracy_score(y_test, predictions))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import plot_tree



# Initialize the Random Forest model
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)

'''f = plt.figure(figsize=(12,12))
plot_tree(RandomForestClassifier, filled=True, rounded=True, 
          feature_names=["Label"], 
          class_names=['sentiment_score', '7_day_avg', '30_day_avg', 'daily_return', 'volatility_30'])
'''
# Predictions
rf_predictions = rf_model.predict(X_test)

# Evaluate the model
print("Random Forest Accuracy:", accuracy_score(y_test, rf_predictions))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

# Initialize the Gradient Boosting model
gb_model = GradientBoostingClassifier()
gb_model.fit(X_train, y_train)

# Predictions
gb_predictions = gb_model.predict(X_test)

# Evaluate the model
print("Gradient Boosting Accuracy:", accuracy_score(y_test, gb_predictions))
```
