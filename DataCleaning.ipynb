```python
df = dataClean
```

```python
top_columns = [f'Top{i}' for i in range(1, 26)]

# Function to combine lists element-wise into a single string
def combine_lists(row):
    combined_text = []
    for item in row:
        if isinstance(item, list):  # Check if item is a list
            combined_text.extend(item)
    return ' '.join(combined_text)

# Apply the combine_lists function row-wise across selected columns
df['combined_text'] = df.apply(lambda row: combine_lists([row[col] for col in top_columns]), axis=1)

df.head()
```

```python
#pip install textblob
from textblob import TextBlob

def get_sentiment(text):
  sentiment = TextBlob(text).sentiment.polarity
  return sentiment


df['sentiment_score'] = df['combined_text'].apply(get_sentiment)
df['sentiment_score'].head()
df.head()
```
